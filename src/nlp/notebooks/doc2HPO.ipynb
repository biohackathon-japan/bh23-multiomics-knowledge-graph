```python
base_url = 'https://doc2hpo.wglab.org'
```

```python
def show(r):
    print(json.dumps(r.json(), indent=4))
```

```python
# for test purpose.
url = f"{base_url}/version"
r = requests.post(url)
show(r)
# {u'ncbo': None, u'java': u'1.8.0_191', u'tomcat': u'8.5.35', u'doc2hpo': u'1.21.0', u'metamaplite': u'metamaplite-3.6.2rc3.jar', u'metamap': u'2016v2'}
```

```python
# for string-based match. faster. 
url = f"{base_url}/parse/acdat"
query = {
    "note": text,
    "negex": True
}
# headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
r = requests.post(url,json = query)
show(r)
```

```python
# with negation detection enabled.
url = f"{base_url}/parse/acdat"
query = {
    "note": text,
    "negex": True
}
# headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
r = requests.post(url,json = query)
show(r)
```

```python
# using metamap lite. Much faster than Original Metamap.
url = f"{base_url}/parse/metamaplite"
query = {
    "note": text,
    "negex": True
}
# headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
r = requests.post(url,json = query)
show(r)
```

```python
# using ncbo annotator - recommended if single file is large.
url = f"{base_url}/parse/ncbo"
query = {
    "note": text,
    "negex": True
}
# headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
r = requests.post(url,json = query)
show(r)
```
